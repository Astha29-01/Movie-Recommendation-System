```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
# Merge the Two Datasets
movies = movies.merge(credits, on='title')
movies.head()
```

```python
# Handle Missing Values
movies.isnull().sum()
movies.dropna(inplace=True)
```

```python
# Convert Text Columns
# The columns genres, keywords, cast, crew are in stringified JSON format.
import ast  # Abstract Syntax Tree

def convert(text):
    # Check if the input is already a list
    if isinstance(text, list):
        # Make sure each item is a dictionary with a 'name' key
        return [i['name'] for i in text if isinstance(i, dict) and 'name' in i]
    
    # If it's a string, parse it using ast.literal_eval
    try:
        L = []
        parsed = ast.literal_eval(text)
        for i in parsed:
            # Check if i is a dictionary and has a 'name' key
            if isinstance(i, dict) and 'name' in i:
                L.append(i['name'])
        return L
    except (ValueError, SyntaxError):
        # Handle the case where text might be in a different format
        return []

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```
